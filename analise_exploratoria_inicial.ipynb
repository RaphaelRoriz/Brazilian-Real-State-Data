## Colunas com valores null

segundo o autor do dataset os valores null foram tratados para aparecerem como 'normal'

Primeiramente eu salvei os nomes de colunas que possuiam valor 'normal' e em seguida verifiquei nestas colunas quantos linhas possuiam o valor 'normal'.

```python
def get_sum_null_values():
    #segundo o autor do dataset os valores null foram tratados para aparecerem como 'normal'
    columns_with_null_values = data.columns[data.isin(['normal']).sum(axis=0) > 0].tolist()

    #getting the name and total number of null values per column
    columns_with_null_values_sum = {}
    for column_name in columns_with_null_values:
        columns_with_null_values_sum[column_name] = data[column_name].isin(['normal']).sum(axis=0)
    
    return columns_with_null_values_sum
```

```python
get_sum_null_values()
```

### Removendo colunas com mais de 70% de valores nulos

```python
columns_to_drop = data.columns[((data.isin(['normal']).sum(axis=0))/number_of_rows) > 0.7].tolist()
```

```python
columns_to_drop
```

```python
for column_name in columns_to_drop:
    data.drop([column_name],axis=1,inplace=True)
```
